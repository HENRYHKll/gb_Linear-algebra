__1.__ Исследовать на линейную зависимость:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

__Решение__

Заметим, что $f_{4}(x)=f_{3}(x)-f_{2}(x)-f_{1}(x)$, то есть вектор $f_{4}(x)$ — линейная комбинация векторов $f_{1}(x)$, $f_{2}(x)$ и $f_{3}(x)$, из чего можно сделать вывод, что $, f_{4}(x)=x-e^{x}$ и $ f_{2}(x)=1, f_{3}(x)=x+1, f_{1}(x)=e^{x}$ линейно зависимы.

__3.__ Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

__Решение__

Стандартный базис линейного пространства $\mathbb{R}^{3}$ образуют векторы 
$e_{1}=(1, 0, 0)$, $e_{2}=(0, 1, 0)$, $e_{3}=(0, 0, 1)$.
Тогда 

$$x=(2, 3, 5)=(2,0,0)+(0, 3, 0)+(0, 0, 5)=\frac{1}{2}\cdot(1, 0, 0)+3\cdot(0, 1, 0)+1\cdot(0, 0, 1)=\frac{1}{2}e_{1}+3e_{2}+e_{3},$$


то есть координатами вектора $x$ в стандартном базисе являются $2$, $3$, $5$.

а) в базисе $1$, $x$, $x^{2}$;

б) в базисе $x^{2}$, $x-1$, $1$.

__Решение__

а) $2, -2, 3$ так как   $2\cdot 1 + \left( { - 2} \right) \cdot x + 3 \cdot {x^2} = 3{x^2} - 2x + 2$

b) $3, -2, 4 $ так как $2 \cdot 1 + 3 \cdot {x^2} - 2\left( {x - 1} \right) + 4 = 3{x^2} - 2x + 2$
